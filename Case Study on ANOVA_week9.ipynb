# ● Check whether all the zones generate the same amount of sales.

We need to check whether all the zones generate the same amount of sales.

We will check whether all the zones generate the same amount of sales or not by checking the equality of means.
Here we are using one way ANOVA

H0=Null Hypothesis
H1=Alternative Hypothesis

In statistical tests, statistical significance is determined by citing an alpha level, or the probability of rejecting the null hypothesis when the null hypothesis is true. In most cases, researchers use an alpha of 0.05, which means that there is a less than 5% chance that the data being tested could have occurred under the null hypothesis.

### H0:All zones generate the same amount of sales


### H1:All zones generate different amount of sales

```python
#input alpha_value
alpha_value=float(input("alpha_value:"))
```

```python
#findinf f_statistic and p_value
f_statistic,p_value=stat.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

```python
f_statistic
```

```python
p_value
```

```python
#comparing the values and printing the result
if p_value<alpha_value:
    print('We reject the null hypothesis since p_value<alpha value i.e ',p_value,"<",alpha_value)
else:
    print('We accept the null hypothesis since p_value>alpha value i.e ',p_value,">",alpha_value)
```

So we can conclude that all zones generate different amount of sales,i.e, atleast one zone generate sales different from others
