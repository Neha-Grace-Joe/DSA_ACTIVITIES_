```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('diabetes_model.csv')
```

```python
pd.DataFrame(data)
```

# 1.count plot of outcome

```python
import warnings
warnings.filterwarnings("ignore")
sns.countplot(data['Outcome'])
```

# 2. Effect of age on outcome using box plot

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Outcome',y='Age',data=data).set_title(" Effect of age on outcome ")
```

# 3.chances of getting diabetes if it is common in family, using the Diabetes   Pedigree Function 

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
plt.title('Chances of getting diabetes,using the DPF',fontsize=15)
plt.xlabel('outcome')
plt.ylabel('Diabetes Pedigree Function')
```

# 4.relationship with Insulin & Glucose

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data['Insulin'],data['Glucose'],hue=data['Outcome'])
plt.title('Relationship with Insulin and Glucose',fontsize=20)
plt.xlabel('insulin',fontsize=20)
plt.ylabel('Glucose',fontsize=20)
warnings.filterwarnings("ignore")
```

```python
data['Insulin'].corr(data['Glucose'])
```

# 5.heatmap on Correlation of features in the data

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True)
```
