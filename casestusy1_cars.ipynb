```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# 1. Reading the dataset to the python environment¶

```python
data=pd.read_csv('cars_data.csv')
data
```

# 2. Checking for the null values present in the dataset

```python
data.isna()
```

```python
#counting the null values in the data set

data.isna().sum()
```

# 3. Ploting a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(6,6))
plt.title('Male vs Female buyers participated in the Sales')
data['Buyer Gender'].value_counts().plot(kind='bar',color=['blue','red'])
plt.xlabel('Gender')
plt.ylabel('count')
plt.grid()
```

# 4. Finding the top 5 cars based on their sales price.

```python
data1=data.nlargest(5, 'Sale Price')
data1
```

```python
top5= data1[["Make", "Model","Sale Price"]]
top5
```

# 5. Find the least 5 cars based on their Resell price.

```python
data2=data.nsmallest(5,"Resell Price")
data2
```

```python
least5= data2[["Make", "Model","Resell Price"]]
least5
```
